```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/student_data.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure()
sns.histplot(df["final_score"], bins=20)
plt.title("Final Score Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```

```python
sns.scatterplot(x="study_hours", y="final_score", data=df)
plt.title("Study Hours vs Final Score")
plt.show()
```

```python
sns.scatterplot(x="attendance", y="final_score", data=df)
plt.title("Attendance vs Final Score")
plt.show()
```

```python
sns.scatterplot(x="internet_usage", y="final_score", data=df)
plt.title("Internet Usage vs Final Score")
plt.show()
```

```python
sns.boxplot(x="extracurricular", y="final_score", data=df)
plt.show()
```
